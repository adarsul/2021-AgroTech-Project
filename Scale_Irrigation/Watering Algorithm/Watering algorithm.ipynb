# The Watering Algorithm

In this expirement we collected data of weight, temeprature and humidity to analyse the change in a pot's weight along time. Using this code we monitored the weight of the plant and sent data to thingspeak. We used a common reed planted in a pot due to it's high water consumption which allowed to collect more data of the change over time. The plant was hand-watered in irregular times (unfortunately) but the graph provided many insights.

To collect data of weight alone use [this code](https://github.com/adarsul/2021-AgroTech-Project/blob/main/Scale_Irrigation/Watering%20Algorithm/Scale_exp.ino)




The data for the graph was collected through thingspeak server and refined in excel. We filtered all readings that were lower than 2400 g (lowest weight was 2500g) and all observations that had a difference bigger than 100g to the following one. The night time of each day was colored blue for easier obseration.

As we can see in the graph, the weight is always highest after watering. At night the weight decreases gradually and the steepest decrease is everyday at midday. 

The weight of the pot could be divided to 3 components:

$ m_c = $ The constant weight which contains the pot and the soil (assuming these values do not change)

$ m_w = $ The weight of water.

$ m_p = $ The weight of the plant.

Therfore $m_{total} = m_c + m_w + m_p$

If a fixed known amount of water is added each time at the same minimum weight, we can conclude that if the weight becomes higher then its due to plant growth.

Hence, when the weight of the pot, soil and change in water are constant,

If $m_{total} > m_{max} $ 
then  $ m_{total} - m_{max} = Growth $

In order for the watering to stay updated, we will assign the current weight to maximum weight and adjust the minimum weight by the increase in mass. 


## VPD

**Vapor Pressure Deficit** is a value that describes the difference between the amount of moisture in the air and the amount of moisture it could hold when saturated. Our interest in VPD is because it has nearly a straight line relationship with evapotranspiration of plants.

on The 4th day of the expirement we added a BME280 climate sensor that measures temperature and humidity which can be used to calculate VPD. For more about the calcutation of VPD [go here](https://pulsegrow.com/blogs/learn/vpd) or just use [our code](https://github.com/adarsul/2021-AgroTech-Project/blob/main/Scale_Irrigation/Watering%20Algorithm/Scale_and_BME.ino).  Notice that you might need to change the values for slope and zero, as well as entering your wifi details and thingspeak channel.
